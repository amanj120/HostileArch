```python
import requests
from bs4 import BeautifulSoup
import csv
import re
```

```python
list_of_bad_words =[]
URL = "https://www.noswearing.com/dictionary/1"
r = requests.get(URL)
# print (r)
soup = BeautifulSoup(r.content, 'html5lib')
# print (soup)
pages = soup.find_all("a", href=re.compile("https://www.noswearing.com/dictionary/"))

for page in pages:
    page = str(page)[9:-7]
    URL = page
    r = requests.get(URL)
    soup = BeautifulSoup(r.content, 'html5lib')
    l = str(soup)
    print(l)
    marker = 0
    while True:
        begin = l.find('</a><b>', marker)
        end = l.find('</b> - ', marker)
        list_of_bad_words.append(l[begin+7:end])
        #print (l[begin+8:end])
        if begin == -1:
            break
        marker = end + 1
#         print (list_of_bad_words)
# print lis
```
